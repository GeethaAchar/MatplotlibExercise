Scatter plots

```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
```

```python
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])

plt.scatter(x,y , color = 'red', marker = 'o', s = 30)
plt.grid(color = 'blue', linestyle = '--', linewidth = 0.5  )
plt.title("Scatter Plot")
plt.show()
```

```python
sizes = np.array([20,50,100,200,500,1000,60,90,10,300,600,800,75])
plt.scatter(x,y , color = 'green', marker = 'o', s = sizes,alpha=0.5)
plt.grid(color = 'blue', linestyle = '--', linewidth = 0.5, alpha = 0.5)
plt.title("Scatter Plot")
plt.show()
```

```python
x = np.random.randint(100, size=(100))
y = np.random.randint(100, size=(100))
colors = np.random.randint(100, size=(100))
sizes = 10 * np.random.randint(100, size=(100))
plt.scatter(x,y,c=colors, s=sizes, alpha=0.5, cmap='nipy_spectral')
```

```python
AgeGroupOne = range(40, 70, 1)                        # X-values ranging from 40 to 69 year old
Purchase1=np.abs(np.random.randn(30) * 20)


AgeGroupTwo=range(20, 40, 1)    # X-values ranging from 20 to 39 years old
Purchase2=np.abs(np.random.randn(20) * 40)
#plt.scatter(x-AgeGroup,y-purchase)
plt.scatter(AgeGroupOne,Purchase1, color = 'red', label="Age 40-69", marker = 'o', s = 100,alpha=0.6)
plt.scatter(AgeGroupTwo,Purchase2, color = 'blue', label="Age 20-39", marker = 'o', s = 100,alpha=0.6)
plt.legend()
plt.grid(color = 'blue', linestyle = '--', linewidth = 0.5  )
plt.title("Purchases by Age")

plt.xlabel("Age")
plt.ylabel("Purchase")  
plt.show()
```

```python
rng = np.random.RandomState(0) # Creates a random number generator (RNG) object with a specific seed value of 0. This ensures reproducibility of the random numbers generated.
x = rng.randn(100) # Generates an array x of 100 random numbers drawn from a standard normal distribution (mean=0, variance=1).
y = rng.randn(100) # Generates an array y of 100 random numbers drawn from a standard normal distribution (mean=0, variance=1).

colors = rng.rand(100) #Generates an array colors of 100 random numbers between 0 and 1.
sizes = 1000 * rng.rand(100) #Generates an array sizes of 100 random numbers between 0 and 1000.
plt.figure(figsize=(10,8))
plt.scatter(x, y, c=colors, s=sizes, alpha=0.3, cmap='RdYlGn')   
plt.colorbar()
plt.grid(color = 'blue', linestyle = '--', linewidth = 0.3  )
plt.xlabel('Random data X')
plt.ylabel('Random data Y')
plt.title("Scatter Plot")
plt.show()
```
