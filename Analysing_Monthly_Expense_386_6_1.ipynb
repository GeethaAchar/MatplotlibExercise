```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
```

Expense Categories: ['Groceries', 'Utilities', 'Transportation,' 'Dining Out,' and 'Entertainment']
Amount Spent: [500, 300, 200, 400, and 250]


```python
ExpenseCategories=['Groceries', 'Utilities', 'Transportation','Dining Out','Entertainment']
AmountSpent=[500, 300, 200, 400, 250]
colors = ['blue', 'yellow','green', 'red', 'purple']
plt.barh(ExpenseCategories, AmountSpent, color=colors)
plt.xlabel('Amount Spent')
plt.ylabel('Expense Categories')
plt.title('Monthly Expense Distribution')
plt.show()
```
