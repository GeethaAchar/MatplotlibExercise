```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

## Basic Lines plots
-time seies data
-trends and patterns in data
-observe linear realtionships

```python
# State based data
xpoint=np.array([1,2,3,4,5])
ypoint=np.array([1,4,9,16,25])
plt.plot(xpoint,ypoint)
plt.show()
```

```python
# object based data
fig,ax=plt.subplots()
ax.plot(xpoint,ypoint)
ax.set(xlabel='x axis',ylabel='y axis',title='Line plot')   
plt.show()
```
