```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
CoursesName = ['C Course', 'C++ Course', 'Java Course', 'Python Course']
Student = [90, 60, 130, 190]
plt.bar(CoursesName, Student,color='#00FFFF',width=0.5)
plt.grid(True,axis='y')
plt.xlabel('Courses Name')
plt.ylabel('Students')
plt.title('Number of Students per Course')
plt.show()
```

```python
plt.barh(CoursesName, Student,color='#00FFFF',height=0.5,zorder=2)
plt.grid(zorder=-1)
plt.xlabel('Students')
plt.ylabel('Courses Name')
plt.title('Number of Students per Course')
plt.show()
```

```python
#Stacked bar chart
male = [90, 60, 130, 600]
female = [100, 30, 160, 690]
plt.bar(CoursesName,male,width=0.5,color='r',label='Male')
plt.bar(CoursesName,female,width=0.5,color='b',label='Female',bottom=male) 
plt.xlabel('Courses Name')
plt.ylabel('Students')
plt.title('Number of Students per Course') 
plt.legend()
plt.show()
```

```python
years = ['2016', '2017', '2018', '2019']
rice_price = [57, 50, 47, 30]
wheat_price = [43, 50, 53, 70]
plt.figure(figsize=(15,10),facecolor='paleturquoise',dpi=100)
plt.bar(years,rice_price,width=0.3,color='#49D48E',label='Rice',zorder=2)
plt.bar(years,wheat_price,width=0.3,color='#49D4D4',label  ='Wheat',bottom=rice_price,zorder=2)
plt.xlabel('Years')
plt.ylabel('Price')
plt.title('Price of Rice and Wheat')
plt.grid(zorder=-1)
plt.legend(loc='upper left')
plt.show()
```

```python
mustard =  np.array([10, 16, 13, 15])
plt.figure(figsize=(12,8),facecolor='paleturquoise',dpi=100)  
plt.bar(years,rice_price,width=0.3,label='Rice',zorder=2)
plt.bar(years,wheat_price,width=0.3,color='#49D4D4',label  ='Wheat',bottom=rice_price,zorder=2)
plt.bar(years, mustard, width=0.3, color="#7466B4", label='Mustard',
        bottom=np.array(rice_price) + np.array(wheat_price))
plt.legend(loc='upper left')
plt.xlabel('Years')
plt.ylabel('Price')
plt.title('Price of Rice, Wheat and Mustard in 2016-2019')
plt.grid(zorder=-1)
plt.show()
```

#clustered column charts

```python
CoursesName = ['C Course', 'C++ Course', 'Java Course', 'Python Course']
StudentGroup1 = [90, 60, 130, 600]
StudentGroup2 = [100, 30, 160, 690]
bar_width=0.35
x=np.arange(len(CoursesName))   

plt.bar(x-bar_width/2, StudentGroup1,color="#F1EE32",width=0.35,label='Group 1')
plt.bar(x+bar_width/2, StudentGroup2,color="#0F99E9",width=0.35,label='Group 2')    
plt.xticks(x,CoursesName,rotation=45)
plt.xlabel('Courses Name')
plt.ylabel('Students')
plt.title('Number of Students per Course')
plt.legend()
plt.show()
```

```python
data = {'CoursesName': ['C Programming Course', 'C++ Programming Course', 'Java Programming Course', 'Python Programming Course'],
        'Student Group 1': [90, 60, 130, 600],
        'Student Group 2': [100, 30, 160, 690]}
df = pd.DataFrame(data)
df  
```

```python
plt.figure(figsize=(15,10),facecolor='paleturquoise',dpi=100) 
bar_width=0.3
x=np.arange(len(df['CoursesName']))


plt.bar(x-bar_width/2, df['Student Group 1'],width=0.3,color='#49D48E',label='Group 1',zorder=2)
plt.bar(x+bar_width/2, df['Student Group 2'],width=0.3,color="#434483",label  ='Group 2',zorder=2)
plt.xlabel('Courses Name')  
plt.ylabel('Students')  
plt.title('Number of Students per Course')
plt.grid(zorder=-1)
plt.legend(loc='upper left')
plt.xticks(x,df['CoursesName'],rotation=45)
plt.show()  
```
