```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
xpoint=np.array([1,8])
ypoint=np.array([3,10])
plt.plot(xpoint,ypoint)
plt.show()
```

```python
plt.plot(xpoint,ypoint,linestyle='--')
plt.show()
```

```python
plt.plot(xpoint,ypoint,linestyle='dashdot',color='blue',linewidth=3)
plt.show
```

```python
fig,ax=plt.subplots()
ax.plot(xpoint,ypoint,linestyle=':',color='green',marker='o')
ax.set(xlabel='x axis',ylabel='y axis',title='Line plot')   
plt.show()
```
