```python
df.loc[9, 'Name'] = 'Unknown'
age_median=df['Age'].median().round(1).astype(int)
df['Age'].fillna(age_median, inplace=True)
weight_median=df['Weight'].mean().round(2)
df['Weight'].fillna(weight_median, inplace=True)
salary_median=df['Salary'].median().round(2)
df['Salary'].fillna(salary_median, inplace=True)
df
```

```python
df=df.sort_values(by='Age', ascending=True)
df
```

```python
plt.figure(figsize=(10, 10),facecolor='paleturquoise',dpi=100)
plt.plot(df['Age'], df['Salary'], 'ro-')
plt.xlabel('Age')
plt.ylabel('Salary')
plt.title('Line graph of Salary vs Age',fontweight='bold', fontsize=15)
plt.xlabel('Age')
plt.ylabel('Salary')
plt.grid(color='red', alpha=0.5, linestyle='dashed', linewidth=0.5)
plt.show()
```
